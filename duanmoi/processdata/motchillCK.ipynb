```python
import  pandas as pd
import numpy as np
```

```python
column_names = [
    "TenPhim", "TrangThai", "DaoDien", "ThoiLuong", "PhuDe",
    "NamSX", "QuocGia", "TheLoai", "MoTa", "ĐanhGia",
    "ĐoPhanGiai", "LuotXem", "LinkPhim"
]
df = pd.read_csv('duanmoi/spiders/csvduanmoi.csv', header=0, names=column_names, delimiter='$', encoding='utf-8')
df
```

```python
# Xóa cột 'LinkPhim'
df = df.drop(columns=['LinkPhim'])

df.head(5)
```

```python
# Tách cột Đánh giá thành 2 cột mới
df[['rating', 'vote']] = df['ĐanhGia'].str.extract(r'\((\d+\.\d+)\s+sao\s+/\s+(\d+)\s+đánh giá\)')

# Xóa cột Đánh giá gốc nếu không cần nữa
df.drop(columns=['ĐanhGia'], inplace=True)

df
```

```python
# 
df['rating'] = df['rating'].astype('float')  # Chuyển đổi sang Int64 cho phép NaN

# Chuyển đổi cột namsx thành int, giả sử namsx đã không có chữ nào
df['vote'] = df['vote'].astype('float')

df['LuotXem'] = df['LuotXem'].str.replace('lượt', '').str.replace(',', '').astype(float)
df.dtypes
```

```python
# Import thư viện
import pandas as pd
from pymongo import MongoClient

# Kết nối đến MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Chọn cơ sở dữ liệu và collection
db = client['dbmotchill']
collection = db['tblmotchill_new']

# Chuyển DataFrame thành danh sách các dictionary
data_dict = df.to_dict(orient="records")

# Chèn dữ liệu vào MongoDB
collection.insert_many(data_dict)

# Kiểm tra dữ liệu đã lưu
for document in collection.find():
    print(document)
```
