```python
import  pandas as pd
import numpy as np
from pymongo import MongoClient
import uuid
# chụp
```

```python
# Kết nối tới MongoDB
client = MongoClient('mongodb://localhost:27017/')  # Thay đổi URL nếu cần
db = client["dbmotchill"]  # Tên cơ sở dữ liệu

# Giả sử bạn có một collection tên là 'your_collection_name'
collection = db["tblmotchill_new"]  # Thay đổi tên collection
# Lấy tất cả tài liệu từ collection
all_documents = list(collection.find())
# Chuyển đổi danh sách tài liệu thành DataFrame
df = pd.DataFrame(all_documents)
# Đóng kết nối
client.close()

df
# chụp
```

```python
# Thay thế giá trị NaN trong cột '_id' bằng số tự động từ 1
df['_id'] = pd.to_numeric(df['_id'], errors='coerce')  # Chuyển đổi cột _id thành kiểu số
df['_id'] = df['_id'].fillna(pd.Series(range(1, len(df) + 1)))  # Điền NaN bằng các số tự động từ 1 trở đi

# Hiển thị kết quả
df.head(5)
#chụp
```
