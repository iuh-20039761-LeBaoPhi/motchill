```python
import  pandas as pd
import numpy as np
from pymongo import MongoClient
import uuid
```

```python
# Kết nối tới MongoDB
client = MongoClient('mongodb://localhost:27017/')  # Thay đổi URL nếu cần
db = client["dbmotchill"]  # Tên cơ sở dữ liệu

# Giả sử bạn có một collection tên là 'your_collection_name'
collection = db["tblmotchill"]  # Thay đổi tên collection
# Lấy tất cả tài liệu từ collection
all_documents = list(collection.find())
# Chuyển đổi danh sách tài liệu thành DataFrame
df = pd.DataFrame(all_documents)
# Đóng kết nối
client.close()

df
```

```python
#kiểm tra số cột, dòng
df.shape
```

```python
# Kiểm tra kiểu dữ liệu
df.info()
```

```python
# Xóa cột '_id' và 'courseUrl'
df = df.drop(columns=['courseUrl'])

df.head(5)
```

```python
# Tách cột Đánh giá thành 2 cột mới
df[['rating', 'vote']] = df['danhgia'].str.extract(r'\((\d+\.\d+)\s+sao\s+/\s+(\d+)\s+đánh giá\)')

# Xóa cột Đánh giá gốc nếu không cần nữa
df.drop(columns=['danhgia'], inplace=True)

df
```

```python
# 
df['rating'] = df['rating'].astype('float')  # Chuyển đổi sang Int64 cho phép NaN

# Chuyển đổi cột namsx thành int, giả sử namsx đã không có chữ nào
df['vote'] = df['vote'].astype('float')

df['luotxem'] = df['luotxem'].str.replace('lượt', '').str.replace(',', '').astype(float)
df.dtypes
```

```python
df['namsx'] = df['namsx'].replace('Đang cập nhật...', 0)
# Chuyển đổi cột _id thành UUID mà không mất dữ liệu
df['_id'] = pd.to_numeric(df['_id'], errors='coerce')  # Chuyển đổi cột _id thành kiểu số, giá trị không thể chuyển đổi sẽ thay thế bằng NaN

# Thay thế NaN bằng UUID (Universally Unique Identifier) mới được tạo ngẫu nhiên
df['_id'] = df['_id'].apply(lambda x: uuid.uuid4() if pd.isnull(x) else x)
df.head(5)
```

```python
#Đổi tên cột
df.rename(columns={'_id':'id_phim', 'coursename':'tenphim'},inplace=True)
df
```
